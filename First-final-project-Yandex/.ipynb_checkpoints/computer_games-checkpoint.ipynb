```python
games.columns = ['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'critic_score', 'user_score', 'rating']
```

Однако, в условиях срока сдачи работы "вчера", а также, что игры с пропусками имеют невысокие продажи (какие-то близки к 0), их проще удалить.

```python
games[games['user_score'] == 'tbd']
```

tbd ('To Be Determined'), означает, что оценок слишком мало, либо их вообще нет. Заменим эти значения на NaN.

```python
import numpy as np
```

```python
games['user_score'] = games['user_score'].replace('tbd', np.nan)
```

#### Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.

```python
games['total_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales'] 
```

```python
games
```

#### Приведем данные к нужным типам.

```python
games['user_score'] = pd.to_numeric(games['user_score'], errors='ignore')
```

```python
games['rating'] = pd.to_numeric(games['rating'], errors='ignore')
```

```python
games['year_of_release'] = games['year_of_release'].astype(int)
```

Данные по оценкам пользователей и рейтингу являются количественными переменными. С помощью функции pd.to_numeric() перевели значения в тип float, а, поскольку в обоих столбцах были пропуски, то данной функции был задан параметр для игнорирования этих значений.  
Данные столбца 'year_of_release' были типа float. Можно было перевести их в тип datetime, однако, нас интересует только год, поэтому эти данные можно выводить в виде целого цисла.

Данный dataframe имеет множество пропусков в столбцах с рейтингом и отзывами. На этапе предобработки удалять их нельзя (это половина всех данных), поскольку в дальнейшем планируется исследование количества продаж в зависимости от тех или иных условий.
Заменять такие пропущенные данные на среднее значение или медиану некорректно.  
Также было замечено, что большинство данных выгружено с пропусками сразу во всех 3х столбцах, что может означать ошибку при выгрузке.  
В то же время можно отметить, что данные об оценках и ретинге отсутствуют у игр с маленькмим или же нулевыми продажами. Заменять такие значения тоже нельзя, мы не можем предугадывать оценки.
```python
group_platform_ps2 = games[games['platform'] == 'PS2']
```

```python
group_platform_x360 = games[games['platform'] == 'X360']
```

```python
data1 = group_platform_ps2.groupby('year_of_release')['total_sales'].sum()
data2 = group_platform_x360.groupby('year_of_release')['total_sales'].sum()
data3 = group_platform_ps3.groupby('year_of_release')['total_sales'].sum()
data4 = group_platform_wii.groupby('year_of_release')['total_sales'].sum()
data5 = group_platform_ds.groupby('year_of_release')['total_sales'].sum()
data6 = group_platform_ps.groupby('year_of_release')['total_sales'].sum()

def plotter():
    plt.figure(figsize=(12,7))
    plt.grid()
    plt.plot(data1, label='PS2')
    plt.plot(data2, label='X360')
    plt.plot(data3, label='PS3')
    plt.plot(data4, label='Wii')
    plt.plot(data5, label='DS')
    plt.plot(data6, label='PS')
    pylab.legend()
    plt.title('Суммарный объем продаж в год')
    plt.show()
    
plotter()
```

Примерно 10 лет существуют новые платформы на рынке, затем они исчезают. Пик продаж приходится примерно на 5й год существования платформы. То есть, платформы, появившиеся в 2012-2013 году будут на пике популярности в 2017. Посмотрим следующие 6 платформ по продажам.

### Проверим гипотезы.
Первая гипотеза для проверки звучит следующим образом: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Поскольку нулевая гипотеза должна звучать таким образом, чтобы быть со знаком равенства, то она будет сформулирована следующим образом: "средние пользовательские рейтинги платформ Xbox One и PC равны". Соотвественно, альтернативная гипотеза будет звучать так: "средние пользовательские рейтинги платформ Xbox One и PC не равны".  
Для проверки этой гипотезы нам нужно получить две выборки: рейтинг пользователей платформ Xbox One и PC. 
Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, применяется метод scipy.stats.ttest_ind (array1, array2, equal_var).  
Для исследования возьмем актуальный период.

```python
u_score_xone = games_actual[games_actual['platform'] == 'XOne']
```

```python
u_score_xone = u_score_xone.dropna(subset=['user_score'])
```

```python
np.var(u_score_xone['user_score'], ddof=1)
```

```python
u_score_pc = games_actual[games_actual['platform'] == 'PC']
```

```python
u_score_pc = u_score_pc.dropna(subset=['user_score'])
```

```python
np.var(u_score_pc['user_score'], ddof=1)
```

```python
results = st.ttest_ind(
    u_score_xone['user_score'], 
    u_score_pc['user_score'], 
    equal_var=False)

alpha = .05

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
u_score_pc['user_score'].mean()
```

```python
u_score_xone['user_score'].mean()
```

#### Вывод: средняя оценка пользователей PC и XOne одинакова.
Поскольку полученное значение p-value говорит нам о том, что вероятность получить такую разницу между средними оценками пользователей PC и XOne равна 60%.

Вторая для проверки гипотеза: средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные.  
Нулевая гипотеза должна звучать так: "средние пользовательские рейтинги жанров Action и Sports равны".  
Для проверки этой гипотезы также применим метод scipy.stats.ttest_ind (array1, array2, equal_var).

```python
u_score_action = games_actual[games_actual['genre'] == 'Action']
```

```python
u_score_action = u_score_action.dropna(subset=['user_score'])
```

```python
u_score_sports = games_actual[games_actual['genre'] == 'Sports']
```

```python
u_score_sports = u_score_sports.dropna(subset=['user_score'])
```

```python
np.var(u_score_action['user_score'], ddof=1)
```

```python
np.var(u_score_sports['user_score'], ddof=1)
```

```python
results = st.ttest_ind(
    u_score_action['user_score'], 
    u_score_sports['user_score'], 
    equal_var=False)

alpha = .05

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Вывод: мы можем достоверно утверждать, что средние пользовательские рейтинги этих двух жанров не равны.
