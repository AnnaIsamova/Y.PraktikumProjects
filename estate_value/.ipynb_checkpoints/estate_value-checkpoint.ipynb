### Предобработка данных

```python
estate['ceiling_height'].unique()
```

Столбец "ceiling_height" содержит в себе аномальные значения: 100, 20, 26 и т.д. Посмотрим распределение данных параметров на гистограмме и решим, какая минимальная и максимальная высота потолков является "пороговой". 

```python
estate.boxplot('ceiling_height', figsize=(5,7))
plt.ylim(0, 5)
```

Из данного графика видно, что значения ниже 2,25 и выше 3.25 можно считать выбросами. Уберем их.

```python
ceiling_height = estate.query('2.25 <= ceiling_height <= 3.25')
```

```python
estate['ceiling_height'] = estate['ceiling_height'].where(estate['ceiling_height'].isin(ceiling_height['ceiling_height']))
```

Заменим все пропуски в занчениях 'floors_total' на 1, так как предполагаем, что это одноэтажное строение.

```python
estate['floors_total'] = estate['floors_total'].fillna(1)
```

Заполним пропущенные значения высоты потолков на основе медианного значения высоты потолков для отдельного по этажности дома.

```python
for i in estate['floors_total'].unique():
    median = estate.loc[estate['floors_total'] == i, 'ceiling_height'].median()
    estate.loc[estate['floors_total'] == i, 'ceiling_height'] =  estate.loc[estate['floors_total'] == i, 'ceiling_height'].fillna(median)
```

```python
estate.loc[estate['ceiling_height'].isna()]
```

Осталось 2 пропущенных значения, поскольку каждое из них только в единичном кол-ве, медиану посчитать не удалось.

```python
estate['balcony'].value_counts()
```

100 или 20 балконов в одной квартире мы не наблюдаем, это хорошо. 

```python
estate['balcony'].isnull().sum()
```

Пропущенных значений много. Возможно, что пропущенные значения - это отсутствие балконов, и об этом не указали в объявлении.  Пусть все пропущенные значения означают, что балконов нет вообще. Заменим пропущенные значения на нули. 

```python
drop_balcony = estate.dropna(subset=['balcony'])
```

```python
estate['balcony'] = (
    estate['balcony']
    .where(estate['balcony']
           .isin(drop_balcony['balcony']), 0.0)
)
```

```python
estate['balcony'].isnull().sum()
```

```python
estate['is_apartment'].unique()
```

```python
estate['is_apartment'].value_counts()
```

```python
is_apartment = estate.loc[estate['is_apartment'] == True]
is_apartment
```

Если посмотреть данные только по апартаментам, нет явных отличительных черт, чтобы сказать, что вот это помещение является апартаментами. Пусть пропущенные значения означают, что помещение не является апартаментами.

```python
estate['is_apartment'] = (
    estate['is_apartment']
    .where(estate['is_apartment']
           .isin(is_apartment['is_apartment']), False)
)
```

```python
estate['is_apartment'].value_counts()
```

```python
estate['is_apartment'] = estate['is_apartment'].astype(bool)
```

Итого, в данном датасете, из всех жилых помещений мы имеет 50 аппартаментов. 

```python
wrong_area = estate.query('total_area < (living_area + kitchen_area)')
wrong_area
```

Из 23699 объявлений в 64 неверно указана общая площадь, либо площадь кухни, либо жилая, так как сумма жилой площади и кухни не может быть больше общей. Как правило, собственники знают общую площадь и площадь кухни, скорее всего ошибка могла быть в заполнении жилой площади(это можно увидеть в строке 21956 и 23202, где видно, что при общей площади 46.00, указана жилая площадь 44.00, и, соответственно, при общей площади 67.92, указана жилая 63.00. Уберем эти ошибочные данные и заполним их ниже.

```python
estate['living_area'] = estate['living_area'].where(~estate['living_area'].isin(wrong_area['living_area']))
estate.query('total_area < (living_area + kitchen_area)')
```

Заполним пропущенные значения по площади кухни.

```python
estate['no_liv_area'] = estate['total_area'] - estate['living_area'] - estate['kitchen_area']
```

Получили площадь коридоров и санузлов. Пропущенные значения этой площади заменим на медианную. Найдем площадь кухни и заменим этим значением все пропущенные в одноименном столбце.

```python
no_liv_area_med = estate['no_liv_area'].median()
estate['no_liv_area'] = estate['no_liv_area'].fillna(no_liv_area_med)
kitchen_area = estate['total_area'] - estate['living_area'] - estate['no_liv_area']
estate['kitchen_area'] = estate['kitchen_area'].fillna(kitchen_area)
```

```python
estate['kitchen_area'].isnull().sum()
```

Осталось 1621 пропущенных занчения, это связано с тем, что есть пропуски в значениях жилой площади
Заполним их медианой.

```python
kitchen_area_med = estate['kitchen_area'].median()
estate['kitchen_area'] = estate['kitchen_area'].fillna(kitchen_area_med)
estate['kitchen_area'].isnull().sum()
```

Рассчитаем жилую площадь:

```python
living_area = estate['total_area'] - ( estate['kitchen_area'] + estate['no_liv_area'])
estate['living_area'] = estate['living_area'].fillna(living_area)
estate['living_area'].isnull().sum()
```

```python
local_name = estate.pivot_table(index = 'locality_name', values='total_images', aggfunc = 'count')
local_name = local_name.sort_values(by='total_images', ascending=False)
local_name.head(10)
```

```python
local_name_mean_sq = estate.pivot_table(index = 'locality_name', values='price_per_sq_m', aggfunc = 'mean')
local_name_mean_sq
```

```python
local_join = local_name_mean_sq.merge(local_name, on='locality_name',how='left')
local_join.sort_values(by='total_images', ascending=False)
```

```python
local_top10 = local_join.query('total_images >= 237')
```

Вывели топ-10 городов по кол-ву объявлений со средней ценой за кв.м.

```python
local_top10.reset_index(level=0, inplace=True)
local_top10.sort_values(by='price_per_sq_m', ascending=False)
```

Самая высокая цена за кв.м в Санкт-Петербурге.

Установить параметры для построения автоматизированной системы для отслеживания аномалий и мошеннических деятельнойстей

  - пропуски в названиях населенных пунктов;

  - аномальные значения высоты потолков;
