Годовую сезонность на представленных данных нам не посчитать, зато можно посмотреть суточную сезонность. На данном графике можно увидеть тренд. Да, действительно, сложно по представленным данным точно утверждать, истинный ли это тренд или годовая сезонность, поскольку имеем данные только за пол года.

```python
plt.figure(figsize=(10,7))
decomposed.trend['2018-04-20':'2018-07-01'].plot(ax=plt.gca(), grid=True)
pylab.legend()
plt.title('Тренд')
plt.show()
```

Можно отметить колебания тренда на отрезках по месяцам. Например, на графике выше можно заметить, что к концу предыдущего - началу следующего месяца отмечается отрицательный или стационарный тренд, а затем идет рост.

```python
plt.figure(figsize=(10,7))
decomposed.resid.plot(ax=plt.gca(), grid=True)
pylab.legend()
plt.title('Остаток')
plt.show()
```

```python
def plotter():
    plt.figure(figsize=(15,7))
    plt.grid()
    plt.plot(data_resample, label='data_resample')
    plt.plot(data_resample['num_orders'].rolling(24).mean(), label='Скользящее среднее')
    plt.plot(data_resample['num_orders'].rolling(24).std(), label='Скользящее стандартное отклонение')
    pylab.legend()
    plt.show()
    
plotter()
```

Если данные слишком сильно меняются, модель не сможет их правильно предсказать, соответственно, RMSE будет высокой. Процесс можно считать стационарным,  если его распределение со временем не меняется (как на графике суточной сезонности). По проведенному тесту Дикки-Фуллера делаем вывод, что ряд уже является стационарным.  
Для обучения модели нам будут важны следующие признаки:
- час (наблюдается суточная сезонность)
- месяц (видим восходящий тренд, возможно это годовая сезонность)
- число (было отмечено изменение трнеда к концу предыдущего-началу следующего месяца)  

Такие признаки, как день недели и год нам не нужны, поскольку не выявлено зависимости количества заказов от определенных дней недели (праздников или выходных), а данные у нас только за половину 2018 года.

# 3. Обучение

**3.1. Создадим признаки**

Итак, выше сделали вывод о том, что для обучения модели нам будут важны следующие "календарные" признаки:
- час (наблюдается суточная сезонность)
- день (по ним можно тоже увидеть изменение тренда к концу-началу месяца)
- месяц (видим восходящий тренд, возможно это годовая сезонность).

Необходимо найти оптимальное количество отстающих значений, они необходимы для того, чтобы модель по предыдущим значениям могла понять, будет ли функция расти или уменьшаться.   
Также необходимо вычислить оптимальное "окно" скользящего среднего, оно задает общий трнед для временного ряда, однако, слишком большое окно может привести к тому, что модель начнет сильнее ошибаться.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

Для формирования признаков, я выбрала lag, равный 24, поскольку таким образом модель сможет делать предсказания на сутки вперед по данным за предыдущие сутки, а rolling mean = 24, поскольку при анализе временного ряда мы обнаружили суточную сезонность (рост заказов примерно с 6 утра до 12 ночи, затем падение).

```python
make_features(data_resample, 24, 24)
```

```python
data_resample = data_resample.dropna()
```

```python
data_resample
```

**3.2 Разделение выборок**

```python
train, test = train_test_split(data_resample, test_size=0.1, shuffle=False)
```

```python
print("Размеры выборок:")
print("Обучающая:",train.shape)
print("Тестовая:",test.shape)
```

**3.3. Определение целевого и зависимых признаков**

```python
x_train = train.drop('num_orders', axis=1)
x_test = test.drop('num_orders', axis=1)
y_train = train['num_orders']
y_test = test['num_orders']
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

Для проведения кросс-валидации нам нужна RMSE, создадим ее.

```python
score = make_scorer(mean_squared_error, squared=False)
```

**3.4. Линейная регрессия**

Поскольку обычная кросс-валидация для подбора гиперпараметров в работе с времнными рядами не подходит, разобьем выборку, на которой будет происходить кросс-валидация для подбора гиперпараметров с помощью TimeSeriesSplit.

```python
reg = LinearRegression()
params = {'fit_intercept':[True,False], 
              'normalize':[False], 
              'copy_X':[True, False]}
my_cv = TimeSeriesSplit(n_splits=3).split(x_train)
grid = GridSearchCV(reg, params, cv=my_cv, scoring=score)
grid.fit(x_train, y_train)
print('Best params:', grid.best_params_)
print('Best score:', grid.best_score_)
```

3.5. **Случайный лес**

```python
param_grid = {
    'bootstrap': [True],
    'max_depth': [8, 9, 10, 11],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
rf_regr = RandomForestRegressor()
my_cv = TimeSeriesSplit(n_splits=3).split(x_train)
rf_grid = GridSearchCV(rf_regr, param_grid, cv=my_cv, verbose=3, scoring=score)
rf_grid.fit(x_train, y_train)
print('Best params:', rf_grid.best_params_)
print('Best score:', rf_grid.best_score_)
```

**3.6 CatBoost**

```python
model = CatBoostRegressor()
grid = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}
my_cv = TimeSeriesSplit(n_splits=3).split(x_train)
CV_model = GridSearchCV(estimator=model, param_grid=grid, cv= my_cv)
CV_model.fit(x_train, y_train)
print(CV_model.best_params_)
```

# 4. Тестирование

```python
lr = LinearRegression(copy_X=True, fit_intercept=True, normalize=False)
lr.fit(x_train, y_train)
prediction = lr.predict(x_test)
plt.figure(figsize=(15, 7))
plt.plot(prediction, "r", label="prediction")
plt.plot(y_test.values, label="actual")
plt.legend(loc="best")
plt.title("Linear regression\n Mean squared error {} orders".format(round(mean_squared_error(y_test, prediction, squared=False))))
plt.grid(True);
```

```python
rf = RandomForestRegressor(bootstrap=True, max_depth=8, max_features=2, 
                           min_samples_split=12, min_samples_leaf=5,
                           n_estimators=100)
rf.fit(x_train, y_train)
prediction_rf = rf.predict(x_test)
plt.figure(figsize=(15, 7))
plt.plot(prediction_rf, "r", label="prediction")
plt.plot(y_test.values, label="actual")
plt.legend(loc="best")
plt.title("Random Forest Regressor\n Mean squared error {} orders".format
          (round(mean_squared_error(y_test, prediction_rf, squared=False))))
plt.grid(True);
```

```python
cat_model = CatBoostRegressor(learning_rate=0.03, depth=6, l2_leaf_reg=1)
cat_model.fit(x_train, y_train)
cat_prediction = cat_model.predict(x_test)
plt.figure(figsize=(15, 7))
plt.plot(cat_prediction, "r", label="prediction")
plt.plot(y_test.values, label="actual")
plt.legend(loc="best")
plt.title("CatBoost Regressor\n Mean squared error {} orders".format
          (round(mean_squared_error(y_test, cat_prediction, squared=False))))
plt.grid(True);
```

**Вывод:**  
Лучше всех на подобранных гиперпараметрах показывается себя модель CatBoost, дальше идет линейная регрессия. Линейная регрессия ошибается в основном на больших выбросах. CatBoost также ошибается на тех же самых выбросах, но уже не так сильно.
