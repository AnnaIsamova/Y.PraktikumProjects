Основываясь на гистограмме, можно выделить, например, три категории авто по пробегу:
- маленький с 0 до 60 тысяч
- средний от 60 до 120 тысяч

```python
print('Пропуски:',data['kilometer'].isna().sum())
data = data.dropna(subset=['kilometer'])
print('После удаления:', data['kilometer'].isna().sum())
```

```python
def mileage_group(kilometers):
    if kilometers <= 60000:
        return 'low mileage'
    if  60000 < kilometers <= 120000:
        return 'average mileage'
    else:
        return 'high mileage'
data['mileage'] = data['kilometer'].apply(mileage_group)
```

# 2. Обучение моделей

2.1. Линейная регрессия

```python
%%time
lr = LinearRegression(normalize=True)
lr.fit(x_train, y_train)
linear_predict = lr.predict(x_test)
print('RMSE:', mean_squared_error(y_test, linear_predict) **0.5)
```

2.2. Случайный лес.

```python
%%time
rfr=RandomForestRegressor(random_state=42)
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8]
}
CV_rfr = GridSearchCV(estimator=rfr, param_grid=param_grid, cv= 5, verbose = 3)

CV_rfr.fit(x_train, y_train)
print(CV_rfr.best_params_)
```

[Parallel(n_jobs=1)]: Done 150 out of 150 | elapsed: 38.1min finished  
{'max_depth': 8, 'max_features': 'log2', 'n_estimators': 500}  
CPU times: user 38min 25s, sys: 2.95 s, total: 38min 28s  
Wall time: 38min 30s

Лучшие параметры: max_depth=8, max_features='log2', n_estimators=500

```python
%%time
forest = RandomForestRegressor(max_depth=8, max_features='log2', n_estimators=500, random_state = 42)
forest.fit(x_train, y_train)
forest_predict = forest.predict(x_test)
print('RMSE:', mean_squared_error(y_test, forest_predict)**0.5)
```

2.3. Catboost

```python
%%time
model = CatBoostRegressor()
grid = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9],
        'iterations': [1000, 3000, 5000]}

CV_model = GridSearchCV(estimator=model, param_grid=grid, cv= 5)

CV_model.fit(x_train, y_train)
print(CV_model.best_params_)
```

{'depth': 10, 'iterations': 5000, 'l2_leaf_reg': 3, 'learning_rate': 0.03}  
CPU times: user 12h 10min 32s, sys: 57min 2s, total: 13h 7min 35s  
Wall time: 7h 20min 31s

```python
%%time
cat = CatBoostRegressor(objective = "RMSE", iterations=5000, depth=10, l2_leaf_reg=3, learning_rate=0.03)
cat.fit(x_train, y_train)
cat_prediction = cat.predict(x_test)
print("RMSE:", mean_squared_error(y_test, cat_prediction)**0.5)
```

2.3. LightBoost

```python
%%time
lgb_reg = lgb.LGBMRegressor(num_boost_round=2000, learning_rate=0.01)
param_grid = {
    'num_leaves': [31, 127],
    'reg_alpha': [0.1, 0.5],
    'min_data_in_leaf': [30, 50, 100, 300, 400]
    }

gridSearchCV = GridSearchCV(estimator = lgb_reg, 
    param_grid = param_grid, 
    cv=5, verbose=2)

gridSearchCV.fit(x_train,y_train)

print(gridSearchCV.best_params_)
```

{'min_data_in_leaf': 30, 'num_leaves': 127, 'reg_alpha': 0.1}  
CPU times: user 1h 1min 7s, sys: 1min 26s, total: 1h 2min 34s  
Wall time: 31min 37s

```python
%%time
lgb_train = lgb.Dataset(x_train, label=y_train)
params = {
    'min_data_in_leaf': 30,
    'num_leaves': 127,
    'reg_alpha': 0.1,
}
gbm = lgb.train(params, lgb_train)

y_pred = gbm.predict(x_test, num_iteration=gbm.best_iteration)
print('RMSE', mean_squared_error(y_test, y_pred)**0.5)
```
